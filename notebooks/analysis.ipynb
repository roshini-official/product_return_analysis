# Step 3: Create Return Flag
We define canceled and unavailable orders as returned.

```python
# Create return flag column
orders['return_flag'] = orders['order_status'].apply(
    lambda x: 1 if x in ['canceled', 'unavailable'] else 0
)

# Check return distribution
orders['return_flag'].value_counts()
```

# Add a new cell and write:
len(orders) → counts total rows

.sum() → adds all 1’s in return_flag

Divide returned by total

Multiply by 100 → percentage

```python
# Total number of orders
total_orders = len(orders)

# Total returned orders
returned_orders = orders['return_flag'].sum()

# Calculate return rate
return_rate = (returned_orders / total_orders) * 100

print("Total Orders:", total_orders)
print("Returned Orders:", returned_orders)
print("Return Rate:", round(return_rate, 2), "%")
```

# Step 1: Merge Orders and Order Items
order_items contains product_id

orders contains return_flag

We join using order_id

Now each product row knows whether it was returned

This is called data integration.

```python
# Merge orders with order_items
merged_data = order_items.merge(
    orders[['order_id', 'return_flag']],
    on='order_id',
    how='left'
)

print("Merged successfully!")
merged_data.head()
```

```python
merged_data.tail()
```

# Step 2: Calculate Return Rate Per Product

```python
# Group by product_id
product_return = merged_data.groupby('product_id').agg(total_orders=('order_id', 'count'),total_returns=('return_flag', 'sum')).reset_index()

# Calculate return rate
product_return['return_rate_%'] = (product_return['total_returns'] / product_return['total_orders']) * 100

# Sort by highest return rate
product_return = product_return.sort_values(by='return_rate_%',ascending=False)

product_return.head(90)
```

# Step 3: Filter Products With Minimum Orders

```python
# Keep products with at least 10 orders
filtered_products = product_return[
    product_return['total_orders'] >= 10
]

# Sort again
filtered_products = filtered_products.sort_values(
    by='return_rate_%',
    ascending=False
)

filtered_products.head(10)
```

# Add Product Category

```python
# Merge with products dataset to get category
product_analysis = filtered_products.merge(
    products[['product_id', 'product_category_name']],
    on='product_id',
    how='left'
)

product_analysis.head(10)
```

# 5. Monthly Return Rate Trend
Logic

This shows how returns change over time.

```python
orders['order_purchase_timestamp'] = pd.to_datetime(
    orders['order_purchase_timestamp']
)

orders['year_month'] = orders['order_purchase_timestamp'].dt.to_period('M').astype(str)

monthly_returns = orders.groupby('year_month').agg(
    total_orders=('order_id','count'),
    total_returns=('return_flag','sum')
).reset_index()

monthly_returns['return_rate_%'] = (
    monthly_returns['total_returns'] /
    monthly_returns['total_orders']
) * 100

plt.figure(figsize=(10,5))

sns.lineplot(
    data=monthly_returns,
    x='year_month',
    y='return_rate_%',
    marker='o'
)

plt.xticks(rotation=45)

plt.title("Monthly Return Rate Over Time")
plt.xlabel("Month")
plt.ylabel("Return Rate (%)")
plt.savefig('outputs/monthly_return_rate.png',dpi=300,bbox_inches='tight')

plt.show()
```

```python
from collections import Counter
import re

# 1. Filter for comments where the order was 'returned' (flag=1)
return_comments = reviews_analysis[
    (reviews_analysis['return_flag'] == 1) & 
    (reviews_analysis['review_comment_message'].notna())
]['review_comment_message']

# 2. Simple function to clean and tokenize text
def get_keywords(text_series):
    all_words = " ".join(text_series).lower()
    # Remove non-alphabetical characters
    words = re.findall(r'\b\w{4,}\b', all_words) # words with 4+ letters
    return words

# 3. Get most common words
words = get_keywords(return_comments)
common_reasons = Counter(words).most_common(10)

print("Top 10 Keywords in Return Comments:")
for word, count in common_reasons:
    print(f"{word}: {count}")
```

```python
# Mapping the Portuguese keywords to English context
translation_map = {
    'produto': 'Product',
    'recebi': 'Received/Not Received',
    'entrega': 'Delivery/Shipping',
    'para': 'For',
    'compra': 'Purchase',
    'ainda': 'Still (Waiting)',
    'pedido': 'Order',
    'loja': 'Store/Seller',
    'depois': 'After (Delay)',
    'prazo': 'Deadline/Lead Time'
}

# Convert your Counter results into a DataFrame for plotting
import pandas as pd

# Using the data you provided
data = {
    'Keyword': ['produto', 'recebi', 'entrega', 'para', 'compra', 'ainda', 'pedido', 'loja', 'depois', 'prazo'],
    'Count': [197, 89, 74, 64, 54, 48, 44, 37, 36, 35]
}

df_reasons = pd.DataFrame(data)

# Apply the translation
df_reasons['Reason (English)'] = df_reasons['Keyword'].map(translation_map)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter out generic words to show actual business reasons
business_reasons = df_reasons[~df_reasons['Reason (English)'].isin(['For', 'Purchase', 'Product'])]

plt.figure(figsize=(10, 6))
sns.barplot(
    data=business_reasons, 
    x='Count', 
    y='Reason (English)', 
    palette='magma'
)

plt.title("Top 7 Business Reasons for Returns (English Translated)")
plt.xlabel("Frequency in Review Comments")
plt.ylabel("Reason")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Calculate total loss for each returned item
merged_data['total_item_loss'] = (merged_data['price'] + merged_data['freight_value']) * merged_data['return_flag']

# Calculate total financial loss
total_financial_loss = merged_data['total_item_loss'].sum()

print(f"Total Financial Loss (Canceled/Unavailable): R$ {total_financial_loss:,.2f}")
```

```python
# Create a list of logistics-related keywords we found earlier
logistics_keywords = ['entrega', 'prazo', 'recebi', 'ainda', 'demora']

# Function to check if a review contains a logistics complaint
def is_logistics_issue(text):
    if pd.isna(text): return False
    return any(word in str(text).lower() for word in logistics_keywords)

# Apply to our reviews_analysis dataframe
reviews_analysis['is_logistics_loss'] = reviews_analysis['review_comment_message'].apply(is_logistics_issue)

# Calculate loss specifically from logistics complaints
logistics_loss = reviews_analysis[reviews_analysis['is_logistics_loss'] == True]['price'].sum()

print(f"Financial Loss due to Logistics Issues: R$ {logistics_loss:,.2f}")
print(f"Percentage of total loss driven by Logistics: {(logistics_loss / total_financial_loss) * 100:.1f}%")
```

```python
# 1. Define the current exchange rate (BRL to INR)
brl_to_inr = 17.54

# 2. Calculate Loss in BRL first
merged_data['total_item_loss_brl'] = (merged_data['price'] + merged_data['freight_value']) * merged_data['return_flag']

# 3. Convert to INR
merged_data['total_item_loss_inr'] = merged_data['total_item_loss_brl'] * brl_to_inr

# 4. Calculate Total Financial Loss in INR
total_loss_inr = merged_data['total_item_loss_inr'].sum()

print(f"Total Financial Loss (INR): ₹ {total_loss_inr:,.2f}")
```

```python
# Calculate loss specifically from logistics complaints in INR
logistics_loss_inr = reviews_analysis[reviews_analysis['is_logistics_loss'] == True]['price'].sum() * brl_to_inr

print(f"Financial Loss due to Logistics Issues: ₹ {logistics_loss_inr:,.2f}")
print(f"Percentage of total loss driven by Logistics: {(logistics_loss_inr / total_loss_inr) * 100:.1f}%")
```
